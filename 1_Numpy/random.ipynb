```python
import matplotlib.pyplot as plt
import numpy as np

ys = [30]
# When you set the same seed value and use the same random number generation functions, 
# you will get the same sequence of random numbers each time you run the code. 
np.random.seed(seed=42)
# 0: main
# 0.5: standard deviation
# 500: number of samples
for delta in np.random.normal(0, 0.5, 500):
    ys.append(ys[-1] + delta)
```

```python
plt.plot(ys)
plt.ylabel('Stock Price ($)')
plt.xlabel('Elapsed Time (min)')
```
