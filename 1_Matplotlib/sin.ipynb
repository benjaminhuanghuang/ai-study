```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sin_fxn(my_x):
    my_y = np.sin(my_x)/my_x
    return my_y
```

```python
x = np.linspace(-10, 10, 10000) # start, finish, n points
y = sin_fxn(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x,y)
```
